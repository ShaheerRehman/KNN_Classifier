```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from math import sqrt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("diabetes.csv")
input_df = df.drop("Outcome", axis=1)
output_df = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_df, output_df, test_size=0.1, random_state=42)

X_train = X_train.values.tolist()
X_test = X_test.values.tolist()
y_train = y_train.values.tolist()
y_test = y_test.values.tolist()

predicted_Y = []


distance = 0
dist_outcome_dict = {}


for i in range(len(X_test)):
  for j in range(8):
    distance += (X_train[i][j] - X_test[i][j])**2

  distance = sqrt(distance)
  dist_outcome_dict[distance] = y_train[i]

dist_outcome = sorted(dist_outcome_dict.items())

for each in dist_outcome:
  p_row_label = each[1]
  predicted_Y.append(p_row_label)

print(predicted_Y)

fn=0
tp=0
fp=0
tn=0


for i in range(len(predicted_Y)):
  if predicted_Y[i]==0 and y_test[i]==0:
    fn+=1
  if predicted_Y[i]==1 and y_test[i]==1:
    tp+=1
  if predicted_Y[i]==1 and y_test[i]==0:
    fp+=1
  if predicted_Y[i]==0 and y_test[i]==1:
    tn+=1

print("False Negative:", fn)
print("True Positive:", tp)
print("False Positive:", fp)
print("True Negative:", tn)


Precision = tp / (tp + fp)
print('Precision:', Precision*100)

from sklearn.metrics import recall_score
recall = recall_score(y_test, predicted_Y, average='binary')
print('Recall: %.3f' % (recall*100))

print('Accuracy:', (accuracy_score(y_test, predicted_Y)*100))
```
